### 1. Which teams have dominated IPL across seasons?

```python
team_wins = matches['winner'].value_counts()

sns.barplot(
    x=team_wins.values,
    y=team_wins.index,
    palette="Set1"
)
plt.title('Total IPL Wins by Each Team', fontsize=14)
plt.xlabel('Number of Wins')
plt.ylabel('Teams')
plt.show()
```

Insight: 
The season-wise win distribution shows that **Mumbai Indians and Chennai Super Kings** have been the most dominant franchises in IPL history.  
Other teams such as **Kolkata Knight Riders and Royal Challengers Bengaluru** have shown competitive performance but with less consistency across seasons.

### 2. Does winning the toss actually help win the match?

```python
toss_win_match= matches[matches['toss_winner']== matches['winner']]
ratio= len(toss_win_match)/len(matches)

print("Ratio of matches where toss winner also won the match: {:.2f}%".format(ratio*100))
```

```python
sns.countplot(x='toss_decision', data=matches, hue='winner', palette='Set1')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.title('Toss Decision vs Match Outcome')
plt.legend(title='Match Winner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Insight:
The analysis shows that **winning the toss does not significantly influence the match outcome**, with the win percentage remaining close to 50%.

### 3. Which Venues Favor Chasing vs Defending?

```python
venue_replacements = {
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium',
    'Brabourne Stadium, Mumbai': 'Brabourne Stadium',
    'M Chinnaswamy Stadium, Bengaluru': 'M Chinnaswamy Stadium',
    'Eden Gardens, Kolkata': 'Eden Gardens',
    'Himachal Pradesh Cricket Association Stadium, Dharamshala': 'Himachal Pradesh Cricket Association Stadium',
    'Arun Jaitley Stadium, Delhi': 'Arun Jaitley Stadium'
}
matches['venue'] = matches['venue'].replace(venue_replacements)
```

```python
matches['win_type']= matches.apply(
    lambda x: 'Chasing' if x['result'] == 'wickets' else 'Defending', axis=1
)
venue_strategy= matches.groupby(['venue', 'win_type']).size().unstack()

venue_strategy.sort_values(by='Chasing', ascending=False).head(10).plot(kind='bar', stacked=True, figsize=(12,6), colormap='Set1')

plt.title('Chasing vs Defending Wins by Venue')
plt.xlabel('Venues')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Matches')
plt.show()
```

Insight:
Venue-level analysis reveals that certain stadiums tend to **favor chasing teams**, while others show a more balanced distribution between chasing and defending wins.
Factors such as pitch conditions, boundary size, and dew during night matches can significantly influence the success of chasing teams at specific venues.

### 4. Has First Innings Score Increased Over the Years?

```python
first_innings = deliveries[deliveries['inning'] == 1]
first_innings_runs = first_innings.groupby('match_id')['total_runs'].sum()

matches['first_innings_runs'] = matches['id'].map(first_innings_runs)
season_avg_score = matches.groupby('season')['first_innings_runs'].mean()
```

```python
season_avg_score.plot(marker='o', figsize=(10,5))

plt.title("Average First Innings Score by Season")
plt.ylabel("Runs")
plt.xlabel("Season")

plt.show()
```

Insight:
The trend of average first innings scores indicates a **gradual increase in scoring rates over the years**.  
Modern IPL seasons tend to produce higher scoring matches compared to the early years of the tournament.

### 5. Who are the most consistent top batsmen?

```python
batsman_runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls_faced = deliveries.groupby('batter').size()
strike_rate = (batsman_runs / balls_faced) * 100

batting_stats = pd.DataFrame({
    'runs': batsman_runs,
    'balls': balls_faced,
    'strike_rate': strike_rate
})

qualified_batsmen = batting_stats[batting_stats['balls'] > 1000]
top_batsmen = qualified_batsmen.sort_values('runs', ascending=False).head(10)
```

```python
sns.barplot(x=top_batsmen['runs'], y=top_batsmen.index, palette='Set1')

plt.title("Most Consistent IPL Batsmen")
plt.xlabel("Runs")
plt.ylabel("Batsmen")
plt.show()
```

Insight:
By filtering batsmen who have faced a minimum number of deliveries, we identify players with **both high run totals and strong strike rates**.

### 6. Which bowlers are the most dangerous in death overs?

```python
death_overs = deliveries[deliveries['over'] >= 16]
runs_conceded = death_overs.groupby('bowler')['total_runs'].sum()
balls_bowled = death_overs.groupby('bowler').size()

economy = runs_conceded / (balls_bowled/6)
death_stats = pd.DataFrame({
    'runs': runs_conceded,
    'balls': balls_bowled,
    'economy': economy
})

best_death_bowlers = death_stats.sort_values('economy').head(10)
```

```python
sns.barplot(x=best_death_bowlers['economy'], y=best_death_bowlers.index, palette='Set1')

plt.title("Best Death Over Bowlers (Economy)")
plt.xlabel("Economy Rate")
plt.ylabel("Bowlers")
plt.show()
```

Insight:
The death overs (16-20) analysis highlights bowlers who maintain **low economy rates during this phase**, indicating their ability to control runs under high pressure and perform effective bowling strategies in crucial match moments.

### 9. Which players were the best auction bargains vs most overpaid?

```python
player_runs = deliveries.groupby('batter')['batsman_runs'].sum().reset_index()
player_runs.rename(columns={'batter':'player'}, inplace=True)
```

```python
auction_runs = auction.merge(player_runs, on='player', how='left')
auction_runs['price_crore'] = auction_runs['Amount'] / 10000000
auction_runs['runs_per_crore'] = auction_runs['batsman_runs'] / auction_runs['price_crore']
```

```python
best_bargains = auction_runs.sort_values(
    'runs_per_crore',
    ascending=False
).head(10)
```

```python
sns.barplot(
    x=best_bargains['runs_per_crore'],
    y=best_bargains['player'],
    palette="Set1",
    ci=None
)

plt.title("Best IPL Auction Bargains (Runs per Crore)")
plt.xlabel("Runs per Crore")
plt.ylabel("Players")
plt.show()
```

Insight:
By comparing player performance with their auction price, we can identify players who deliver **exceptional value relative to their cost**.
These players represent the **best auction bargains**, providing high performance while requiring relatively lower investment from their teams.

### 10. Which Team Has the Best Auction Strategy?

```python
team_spending = auction.groupby('Team')['Amount'].sum()
team_runs = auction_runs.groupby('Team')['batsman_runs'].sum()
```

```python
team_value = pd.DataFrame({
    'spending': team_spending,
    'runs': team_runs
})

team_value['spending_crore'] = team_value['spending'] / 10000000
team_value['runs_per_crore'] = team_value['runs'] / team_value['spending_crore']
team_value = team_value.sort_values('runs_per_crore', ascending=False)
```

```python
sns.barplot(
    x=team_value['runs_per_crore'],
    y=team_value.index,
    palette="Set1"
)

plt.title("IPL Team Auction Efficiency")
plt.xlabel("Runs per Crore Spent")
plt.ylabel("Teams")
plt.show()
```

Insight:
Team-level analysis comparing auction spending with player performance highlights franchises that maximize **performance relative to financial investment**.
