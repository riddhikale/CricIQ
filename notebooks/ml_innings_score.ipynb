### First Innings Score Predictor

##### Problem Statement:
The goal of this model is to predict the final first innings score in an IPL match using information available during the powerplay (overs 1-6).
This type of prediction is useful in cricket analytics, where teams estimate their final score early in the innings to adjust their batting strategy.

##### Target Variable:
total_first_innings_score

##### Features:
From deliveries dataset:
powerplay_runs
powerplay_wickets

From matches dataset:
batting_team
bowling_team
venue
season

##### Model Type:
Regression
