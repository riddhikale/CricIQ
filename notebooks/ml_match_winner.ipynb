### Match Winner Predictor

##### Problem Statement:
The goal of this model is to predict the winner of an IPL match based on match conditions such as teams, toss decision, venue, and historical team performance.

##### Target Variable:
`winner` 

We convert the winner into a binary outcome:
Team1 wins → 1
Team2 wins → 0
Features:

##### Features:
From matches dataset:
* toss_winner
* toss_decision
* venue
* team1
* team2
* season

Engineered features:
* team1_win_rate
* team2_win_rate
* team1_avg_score
* team2_avg_score

##### Model Type:
Classification
